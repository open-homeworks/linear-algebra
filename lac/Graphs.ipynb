## Definition of Functions

### Sum of vectors

```python
def sum_vector(a,b):
  len_a = len(a)
  c = v.add(a,b)
  fig = plt.figure(figsize=(10, 10))
  colors = ['#F086C9','#6AE0B0','#F5B878']
 
  if len_a == 2:
    soa = np.array([[0,0,a[0],a[1]], [a[0], a[1], b[0], b[1]], [0, 0, c[0], c[1]]])
    X, Y, U, V = zip(*soa)
    plt.figure()
    ax = plt.gca()
    ax.quiver(X, Y, U, V, angles='xy', scale_units='xy', scale=1, color=colors)
    ax.set_xlim([min(a)-2, max(a)+2])
    ax.set_ylim([min(a)-2, max(a)+2])
    vec1 = np.array([a[0], a[1]])
    vec2 = np.array([b[0], b[1]])
    vec3 = np.array([c[0], c[1]])
    plt.text(max(c)+1, max(c)+0.8, vec1, size=10,ha="center", va="center",bbox=dict(boxstyle="round", color = '#F086C9'))
    plt.text(max(c)+1, max(c), vec2, size=10,ha="center", va="center",bbox=dict(boxstyle="round", color = '#6AE0B0'))
    plt.text(max(c)+1, max(c)-0.8, vec3, size=10,ha="center", va="center",bbox=dict(boxstyle="round", color = '#F5B878'))

  elif len_a == 3:
    soa = np.array([[0, 0, 0, a[0],a[1],a[2]], [a[0], a[1], a[2], b[0], b[1],b[2]], [0, 0, 0, c[0], c[1], c[2]]])
    X, Y,Z, U, V, W = zip(*soa)
    ax = fig.gca(projection='3d')
    ax.quiver(X, Y, Z, U, V, W, color=colors)
    ax.set_xlim([min(a)-2, max(a)+2])
    ax.set_ylim([min(a)-2, max(a)+2])
    ax.set_zlim([min(a)-2, max(a)+2])
    ax.set_ylabel('z')
    vec1 = np.array([a[0], a[1],a[2]])
    vec2 = np.array([b[0], b[1],b[2]])
    vec3 = np.array([c[0], c[1], c[2]])
    ax.text(max(c)+1, max(c)+0.8, max(c)+1, vec1, size=10,ha="center", va="center",bbox=dict(boxstyle="round", color = '#F086C9'))
    ax.text(max(c)+1, max(c), max(c), vec2, size=10,ha="center", va="center",bbox=dict(boxstyle="round", color = '#6AE0B0'))
    ax.text(max(c)+1, max(c)-0.8, max(c)-1, vec3, size=10,ha="center", va="center",bbox=dict(boxstyle="round", color = '#F5B878'))
  
  else:
      raise Exception('The lenght of the vector must be less or equal to 3 to be plotted')

  ax.set_xlabel('x')
  ax.set_ylabel('y')
  plt.title('Addition of two vectors')
  plt.grid()
  plt.draw()
  plt.show()

  return None
```

```python
sum_vector(v1, v2)
```

### Substraction of Vectors

```python
def sub_vector(a,b):
  len_a = len(a)
  c = v.subtract(a,b)
  fig = plt.figure(figsize=(10, 10))
  colors = ['#DE8CC0','#76DADF','#DECC5F']

  if len_a == 2:
    soa = np.array([[0,0,a[0],a[1]], [a[0], a[1], -b[0], -b[1]], [0, 0, c[0], c[1]]])
    X, Y, U, V = zip(*soa)
    ax = plt.gca()
    ax.quiver(X, Y, U, V, angles='xy', scale_units='xy', scale=1, color=colors)
    ax.set_xlim([min(c)-2, max(c)+2])
    ax.set_ylim([min(c)-2, max(c)+2])
    vec1 = np.array([a[0], a[1]])
    vec2 = np.array([b[0], b[1]])
    vec3 = np.array([c[0], c[1]])
    plt.text(max(c)+1, max(c)+0.8, vec1, size=10,ha="center", va="center",bbox=dict(boxstyle="round", color = '#DE8CC0'))
    plt.text(max(c)+1, max(c), vec2, size=10,ha="center", va="center",bbox=dict(boxstyle="round", color = '#76DADF'))
    plt.text(max(c)+1, max(c)-0.8, vec3, size=10,ha="center", va="center",bbox=dict(boxstyle="round", color = '#DECC5F'))

  elif len_a == 3:
    soa = np.array([[0, 0, 0, a[0],a[1],a[2]], [a[0], a[1], a[2], -b[0], -b[1], -b[2]], [0, 0, 0, c[0], c[1], c[2]]])
    X, Y,Z, U, V, W = zip(*soa)
    ax = fig.gca(projection='3d')
    ax.quiver(X, Y, Z, U, V, W, color=colors)
    ax.set_xlim([min(c)-2, max(c)+2])
    ax.set_ylim([min(c)-2, max(c)+2])
    ax.set_zlim([min(c)-2, max(c)+2])
    vec1 = np.array([a[0], a[1],a[2]])
    vec2 = np.array([b[0], b[1],b[2]])
    vec3 = np.array([c[0], c[1], c[2]])
    ax.text(max(c)+1, max(c)+0.8, max(c)+1, vec1, size=10,ha="center", va="center",bbox=dict(boxstyle="round", color = '#DE8CC0'))
    ax.text(max(c)+1, max(c), max(c), vec2, size=10,ha="center", va="center",bbox=dict(boxstyle="round", color = '#76DADF'))
    ax.text(max(c)+1, max(c)-0.8, max(c)-1, vec3, size=10,ha="center", va="center",bbox=dict(boxstyle="round", color = '#DECC5F'))
    ax.set_ylabel('z')
  
  else:
      raise Exception('The lenght of the vector must be less or equal to 3 to be plotted')


  ax.set_xlabel('x')
  ax.set_ylabel('y')
  plt.title('Substraction of two vectors')
  plt.grid()
  plt.draw()
  plt.show()
```

```python
sub_vector(v1,v2)
```

### Cross Product

```python
def cross_product(a,b):

  c = v.cross(a,b)
  colors = ['#E09684','#6E8DE0','#A8E058']
  fig = plt.figure(figsize=(10,10))

  soa = np.array([[0, 0, 0, a[0],a[1],a[2]], [0, 0, 0, b[0], b[1], b[2]], [0, 0, 0, c[0], c[1], c[2]]])
  X, Y, Z, U, V, W = zip(*soa)
  ax = fig.gca(projection='3d')
  ax.quiver(X, Y, Z, U, V, W, color=colors)
  ax.set_xlim([min(c)-2, max(c)+2])
  ax.set_ylim([min(c)-2, max(c)+2])
  ax.set_zlim([min(c)-2, max(c)+2])
  vec1 = np.array([a[0], a[1],a[2]])
  vec2 = np.array([b[0], b[1],b[2]])
  vec3 = np.array([c[0], c[1], c[2]])
  ax.text(max(c)+1, max(c)+0.8, max(c)+1, vec1, size=10,ha="center", va="center",bbox=dict(boxstyle="round", color = '#E09684'))
  ax.text(max(c)+1, max(c), max(c), vec2, size=10,ha="center", va="center",bbox=dict(boxstyle="round", color = '#6E8DE0'))
  ax.text(max(c)+1, max(c)-0.8, max(c)-1, vec3, size=10,ha="center", va="center",bbox=dict(boxstyle="round", color = '#A8E058'))
  ax.set_xlabel('x')
  ax.set_ylabel('y')
  ax.set_zlabel('z')
  plt.title('Cross Product of two vectors')
  plt.grid()
  plt.draw()
  plt.show()

  return c
```

```python
v3 = v.Vector([1,2,3])
v4 = v.Vector([4,5,6])
```

```python
c = cross_product(v3, v4)
```
